```python
import pandas as pd
import os

# Check which files exist in the data folder
print("Files in 'data/' folder:", os.listdir("data"))

# Define the expected file paths
file_paths = {
    "Benin": "data/benin_clean.csv",
    "Sierra Leone": "data/sierraleone_clean.csv",
    "Togo": "data/togo_clean.csv"
}

# Dictionary to store loaded DataFrames
datasets = {}

# Load each CSV
for country, path in file_paths.items():
    if os.path.exists(path):
        datasets[country] = pd.read_csv(path)
        print(f"{country} dataset loaded. Shape: {datasets[country].shape}")
    else:
        print(f"File not found: {path}")

# Optional: preview first few rows of each dataset
for country, df in datasets.items():
    print(f"\n{country} dataset preview:")
    display(df.head())
```

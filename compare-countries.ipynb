```python
import pandas as pd
from IPython.display import display, FileLink

# Dictionary of file paths
file_paths = {
    "Benin": "data/benin_clean.csv",
    "SierraLeone": "data/sierraleone_clean.csv",
    "Togo": "data/togo_clean.csv"
}

# Load, save, and create download links
for country, path in file_paths.items():
    try:
        # Load cleaned dataset
        df = pd.read_csv(path)
        
        # Save to new CSV
        save_name = f"{country.lower().replace(' ', '_')}_clean_final.csv"
        df.to_csv(save_name, index=False)
        
        # Display clickable download link
        print(f"{country} dataset saved as {save_name}:")
        display(FileLink(save_name))
        
    except FileNotFoundError:
        print(f"File not found: {path}")
```
